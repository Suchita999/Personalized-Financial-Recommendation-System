```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')

# Set style
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("Set2")
```

```python
# Define paths (adjust as needed)
data_dir = Path('./data/diary24')
output_dir = Path('./data/output-diary')
output_dir.mkdir(parents=True, exist_ok=True)

quarters = ['241', '242', '243', '244']
quarter_labels = ['Q1 2024', 'Q2 2024', 'Q3 2024', 'Q4 2024']
quarter_abbrev = ['Q1', 'Q2', 'Q3', 'Q4']

print("="*50)
print("QUARTERLY TREND ANALYSIS - CE DIARY SURVEY 2024")
print("="*50)
```

```python
print("\nLoading quarterly data...")

# Load all files by quarter
fmld_quarters = {}
memd_quarters = {}
expd_quarters = {}
dtbd_quarters = {}
dtid_quarters = {}

for i, q in enumerate(quarters):
    print(f"\nLoading Quarter {i+1} (2024)...")
    
    try:
        fmld_quarters[q] = pd.read_csv(data_dir / f'fmld{q}.csv')
        print(f"  ✓ FMLD: {len(fmld_quarters[q]):,} consumer units")
    except:
        print(f"  ✗ FMLD file not found")       
    try:
        memd_quarters[q] = pd.read_csv(data_dir / f'memd{q}.csv')
        print(f"  ✓ MEMD: {len(memd_quarters[q]):,} household members")
    except:
        print(f"  ✗ MEMD file not found")
        
    try:
        expd_quarters[q] = pd.read_csv(data_dir / f'expd{q}.csv')
        print(f"  ✓ EXPD: {len(expd_quarters[q]):,} expenditure records")
    except:
        print(f"  ✗ EXPD file not found")
        
    try:
        dtbd_quarters[q] = pd.read_csv(data_dir / f'dtbd{q}.csv')
        print(f"  ✓ DTBD: {len(dtbd_quarters[q]):,} income records")
    except:
        print(f"  ✗ DTBD file not found")
        
    try:
        dtid_quarters[q] = pd.read_csv(data_dir / f'dtid{q}.csv')
        print(f"  ✓ DTID: {len(dtid_quarters[q]):,} imputed income records")
    except:
        print(f"  ✗ DTID file not found")

print("\n" + "="*80)
print("DATA LOADED - CREATING VISUALIZATIONS")
print("="*80)
```

```python
print("\n1. Creating sample size comparison...")

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Consumer Units per quarter
cu_counts = [len(fmld_quarters[q]) for q in quarters]
colors_cu = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A']

axes[0, 0].bar(quarter_abbrev, cu_counts, color=colors_cu, edgecolor='black', alpha=0.8)
axes[0, 0].set_ylabel('Number of Consumer Units', fontsize=11)
axes[0, 0].set_title('Consumer Units Surveyed by Quarter', fontsize=12, fontweight='bold')
axes[0, 0].grid(axis='y', alpha=0.3)
for i, v in enumerate(cu_counts):
    axes[0, 0].text(i, v, f'{v:,}', ha='center', va='bottom', fontsize=10, fontweight='bold')

# Household members per quarter
member_counts = [len(memd_quarters[q]) for q in quarters]
axes[0, 1].bar(quarter_abbrev, member_counts, color=colors_cu, edgecolor='black', alpha=0.8)
axes[0, 1].set_ylabel('Number of Household Members', fontsize=11)
axes[0, 1].set_title('Total Household Members by Quarter', fontsize=12, fontweight='bold')
axes[0, 1].grid(axis='y', alpha=0.3)
for i, v in enumerate(member_counts):
    axes[0, 1].text(i, v, f'{v:,}', ha='center', va='bottom', fontsize=10, fontweight='bold')

# Average household size by quarter
avg_hh_size = [member_counts[i] / cu_counts[i] for i in range(4)]
axes[1, 0].plot(quarter_abbrev, avg_hh_size, marker='o', linewidth=3, 
                markersize=12, color='#e74c3c', markeredgecolor='black', markeredgewidth=2)
axes[1, 0].fill_between(range(4), avg_hh_size, alpha=0.3, color='#e74c3c')
axes[1, 0].set_ylabel('Average Household Size', fontsize=11)
axes[1, 0].set_title('Average Household Size by Quarter', fontsize=12, fontweight='bold')
axes[1, 0].grid(alpha=0.3)
axes[1, 0].set_ylim([min(avg_hh_size) - 0.1, max(avg_hh_size) + 0.1])
for i, v in enumerate(avg_hh_size):
    axes[1, 0].text(i, v, f'{v:.2f}', ha='center', va='bottom', fontsize=10, fontweight='bold')

# Expenditure records per quarter
expd_counts = [len(expd_quarters[q]) for q in quarters]
axes[1, 1].bar(quarter_abbrev, expd_counts, color=colors_cu, edgecolor='black', alpha=0.8)
axes[1, 1].set_ylabel('Number of Expenditure Records', fontsize=11)
axes[1, 1].set_title('Expenditure Transactions by Quarter', fontsize=12, fontweight='bold')
axes[1, 1].grid(axis='y', alpha=0.3)
for i, v in enumerate(expd_counts):
    axes[1, 1].text(i, v, f'{v:,}', ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.savefig(output_dir / 'quarterly_01_sample_sizes.png', dpi=300, bbox_inches='tight')
print("  ✓ Saved: quarterly_01_sample_sizes.png")
plt.close()
```

```python
# ============================================================================
# VISUALIZATION 2: Expenditure Trends by Quarter
# ============================================================================

print("2. Creating expenditure trend analysis...")

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Total expenditure by quarter
total_expd = []
for q in quarters:
    cost = pd.to_numeric(expd_quarters[q]['COST'], errors='coerce').sum()
    total_expd.append(cost)

axes[0, 0].bar(quarter_abbrev, total_expd, color=colors_cu, edgecolor='black', alpha=0.8)
axes[0, 0].set_ylabel('Total Expenditure ($)', fontsize=11)
axes[0, 0].set_title('Total Expenditures by Quarter', fontsize=12, fontweight='bold')
axes[0, 0].grid(axis='y', alpha=0.3)
for i, v in enumerate(total_expd):
    axes[0, 0].text(i, v, f'${v/1e6:.2f}M', ha='center', va='bottom', fontsize=10, fontweight='bold')

# Average expenditure per CU by quarter
avg_expd_per_cu = [total_expd[i] / cu_counts[i] for i in range(4)]
axes[0, 1].plot(quarter_abbrev, avg_expd_per_cu, marker='o', linewidth=3, 
                markersize=12, color='#27ae60', markeredgecolor='black', markeredgewidth=2)
axes[0, 1].fill_between(range(4), avg_expd_per_cu, alpha=0.3, color='#27ae60')
axes[0, 1].set_ylabel('Average Expenditure per CU ($)', fontsize=11)
axes[0, 1].set_title('Average Weekly Expenditure per Consumer Unit', fontsize=12, fontweight='bold')
axes[0, 1].grid(alpha=0.3)
for i, v in enumerate(avg_expd_per_cu):
    axes[0, 1].text(i, v, f'${v:.2f}', ha='center', va='bottom', fontsize=10, fontweight='bold')

# Median expenditure by quarter
median_expd = []
for q in quarters:
    cost = pd.to_numeric(expd_quarters[q]['COST'], errors='coerce').median()
    median_expd.append(cost)

axes[1, 0].bar(quarter_abbrev, median_expd, color=colors_cu, edgecolor='black', alpha=0.8)
axes[1, 0].set_ylabel('Median Transaction ($)', fontsize=11)
axes[1, 0].set_title('Median Transaction Amount by Quarter', fontsize=12, fontweight='bold')
axes[1, 0].grid(axis='y', alpha=0.3)
for i, v in enumerate(median_expd):
    axes[1, 0].text(i, v, f'${v:.2f}', ha='center', va='bottom', fontsize=10, fontweight='bold')

# Number of transactions per CU by quarter
transactions_per_cu = [expd_counts[i] / cu_counts[i] for i in range(4)]
axes[1, 1].plot(quarter_abbrev, transactions_per_cu, marker='s', linewidth=3, 
                markersize=12, color='#8e44ad', markeredgecolor='black', markeredgewidth=2)
axes[1, 1].fill_between(range(4), transactions_per_cu, alpha=0.3, color='#8e44ad')
axes[1, 1].set_ylabel('Transactions per CU', fontsize=11)
axes[1, 1].set_title('Average Transactions per Consumer Unit', fontsize=12, fontweight='bold')
axes[1, 1].grid(alpha=0.3)
for i, v in enumerate(transactions_per_cu):
    axes[1, 1].text(i, v, f'{v:.1f}', ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.savefig(output_dir / 'quarterly_02_expenditure_trends.png', dpi=300, bbox_inches='tight')
print("  ✓ Saved: quarterly_02_expenditure_trends.png")
plt.close()
```
